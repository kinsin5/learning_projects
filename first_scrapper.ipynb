```python
from bs4 import BeautifulSoup
import requests

url = 'https://pl.wikipedia.org/wiki/Polska#Gospodarka'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
region_table = soup.find_all('table', class_ = 'wikitable')[2]
print(region_table)
```

```python
th_list = region_table.find_all('th')
del th_list[0]
```

```python
header = [x.text.strip() for x in th_list]
header
```
