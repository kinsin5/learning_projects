#### First scrapper out of Polish wiki i wanted to copy this table from this [site](https://pl.wikipedia.org/wiki/Polska#Gospodarka)

 

```python
from bs4 import BeautifulSoup
import requests

url = 'https://pl.wikipedia.org/wiki/Polska#Gospodarka'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
region_table = soup.find_all('table', class_ = 'wikitable')[2]
print(region_table)
```

```python
th_list = region_table.find_all('th')
del th_list[0]
```

```python
header = [x.text.strip() for x in th_list]
header
```

```python
import pandas as pd
df = pd.DataFrame(columns = header)
```

```python
tr_list = region_table.find_all('tr')
del tr_list[0]
```

```python
for row in tr_list:
    td_list = row.find_all('td')
    clean_list = [x.text.strip() for x in td_list]
    
    length = len(df)
    df.loc[length] = clean_list[1:]
```
