```python
###drawing board
def verts():
    for x in range(3):
        print(" ",end="")
        for y in range(3):
            print("-",end="")
    print("")
def horiz(arra):
    str = ""
    for x in range(3):
        str = str + "|"
        str = str + f" {arra[x]} "
    str = str + "|"
    print(str)

def draw(game):
    for x in game:
        verts()
        horiz(x)
    verts()
```

```python
###changing game tiles
def set_game(cords, play_s):
    row = (int)(cords[0]) - 1
    col = (int)(cords[1]) - 1
    print(row, col)
    if play_s == 'p1':
        game[row][col] = "X"
    elif play_s == 'p2':
        game[row][col] = "O"

def inserting_titles():
    counter = 0
    moves = []
    while (counter < 9):
        if counter % 2 == 0:
            player1 = input("P1 - Give your move[row,col]: ")
            if player1 in moves:
                print("It's taken!")
                continue
            moves.append(player1)
            player1 = player1.split(",")
            counter = counter + 1
            set_game(player1, 'p1')
            draw(game)
        if counter != 9 and counter % 2 == 1:
            player2 = input("P2 - Give your move[row,col]: ")
            if player2 in moves:
                print("It's taken!")
                continue
            moves.append(player2)
            player2 = player2.split(",")
            counter = counter + 1
            set_game(player2, 'p2')
            draw(game)       
```

```python
###checking games
def check_row(game):
    won = 'n'
    for x in game:
        first_digit = x[0]
        counter = 0
        for y in x:
            if y == first_digit:
                first_digit = y
                counter = counter + 1
            else:
                break
        if counter == 3:
            print(f"Game has been won by player {x[0]}")
            won = 'y'
            break
    return won

def check_horz(game):
    won = 'n'
    for x in range(3):
        first_digit = game[0][x]
        counter = 0
        for y in range(3):
            if game[y][x] == first_digit:
                first_digit = game[y][x]
                counter = counter + 1
            else:
                break
        if counter == 3:
            print(f"Game has been won by player {game[0][x]}")
            won = 'y'
            break
    return won

def check_diagonal(game):
    p_check = game[1][1]
    if game[0][0] == p_check and game[2][2] == p_check:
        print(f"Game has been won by player {p_check}")
    elif game[0][2] == p_check and game[2][0] == p_check:
        print(f"Game has been won by player {p_check}")
    else:
        print("Nobody won")

def who_won(game):
    if check_row(game) == 'n':
        if check_horz(game) == "n":
            check_diagonal(game)
```

```python
##using all of the functions
game = [[0, 0, 0],[0, 0, 0],[0, 0, 0]]
def tic_tac_toe():
    inserting_titles()
    who_won(game)
tic_tac_toe()
```
